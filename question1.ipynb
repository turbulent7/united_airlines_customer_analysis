```python
final_merged=pd.merge(merged_df,df1,on='call_id',how='left')
final_merged
```

```python
final_merged.isnull().sum()
```

```python
# Convert the string columns to datetime
final_merged['call_end_datetime'] = pd.to_datetime(final_merged['call_end_datetime'])
final_merged['agent_assigned_datetime'] = pd.to_datetime(final_merged['agent_assigned_datetime'])
final_merged['call_start_datetime'] = pd.to_datetime(final_merged['call_start_datetime'])

# Perform subtraction to create the time difference columns
final_merged['Handle_time'] = final_merged['call_end_datetime'] - final_merged['agent_assigned_datetime']
final_merged['Speed_time'] = final_merged['agent_assigned_datetime'] - final_merged['call_start_datetime']

# Convert the time differences to minutes (total_seconds() returns seconds, we divide by 60)
final_merged['Handle_time_minutes'] = final_merged['Handle_time'].dt.total_seconds() / 60
final_merged['Speed_time_minutes'] = final_merged['Speed_time'].dt.total_seconds() / 60

# Display the resulting DataFrame with time in minutes
print(final_merged[['Handle_time_minutes', 'Speed_time_minutes']])
```

```python
final_merged
```

```python
# Calculate the average of multiple columns
average_handle_time = final_merged['Handle_time_minutes'].mean()
average_speed_time = final_merged['Speed_time_minutes'].mean()


print(f"Average Handle Time (in minutes): {round(average_handle_time, 2)}")
print(f"Average Speed Time (in minutes): {round(average_speed_time, 2)}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Group by agent_tone and calculate the average (mean) of Handle_time_minutes
grouped_data = final_merged.groupby('agent_tone').agg({
    'Handle_time_minutes': 'mean',

}).reset_index()
# Plotting the bar plot for Average Handle Time
plt.figure(figsize=(10, 5))
sns.barplot(x='agent_tone', y='Handle_time_minutes', data=grouped_data)
plt.title('Average Handle Time by Agent Tone')
plt.xlabel('Agent Tone')
plt.ylabel('Average Handle Time (in minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python


# Group by agent_tone and calculate the sum of Handle_time_minutes and Speed_time_minutes
grouped_data = final_merged.groupby('agent_tone').agg({
    'Handle_time_minutes': 'sum',
    'Speed_time_minutes': 'sum'
}).reset_index()

# Plotting the bar plot for Handle Time
plt.figure(figsize=(10, 5))
sns.barplot(x='agent_tone', y='Handle_time_minutes', data=grouped_data)
plt.title('Sum of Handle Time by Agent Tone')
plt.xlabel('Agent Tone')
plt.ylabel('Sum of Handle Time (in minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
grouped_data = final_merged.groupby('customer_tone').agg({
    'Handle_time_minutes': 'sum',
}).reset_index()

# Plotting the bar plot for Handle Time
plt.figure(figsize=(10, 5))
sns.barplot(x='customer_tone', y='Handle_time_minutes', data=grouped_data)
plt.title('Sum of Handle Time by Customer Tone')
plt.xlabel('customer_tone')
plt.ylabel('Sum of Handle Time (in minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
grouped_data = final_merged.groupby('customer_tone').agg({
    'Handle_time_minutes': 'mean',
}).reset_index()

# Plotting the bar plot for Handle Time
plt.figure(figsize=(10, 5))
sns.barplot(x='customer_tone', y='Handle_time_minutes', data=grouped_data)
plt.title('mean of Handle Time by Customer Tone')
plt.xlabel('customer_tone')
plt.ylabel('mean of Handle Time (in minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
tone_mapping = {
    'angry': 0,
    'neutral': 1,
    'happy': 2,
    'sad': 3,
    'frustrated': 4,
}

# Map the customer_tone column to numerical values
final_merged['customer_tone_numeric'] = final_merged['customer_tone'].map(tone_mapping)
plt.figure(figsize=(10, 5))
plt.hexbin(x=final_merged['customer_tone_numeric'], y=final_merged['Handle_time_minutes'], gridsize=30, cmap='Blues')
plt.title('Hexbin: Handle Time Distribution by Customer Tone')
plt.xlabel('Customer Tone (Numeric)')
plt.ylabel('Handle Time (in minutes)')
plt.colorbar(label='Count in Bin')
plt.xticks(ticks=list(tone_mapping.values()), labels=list(tone_mapping.keys()), rotation=45)
plt.show()
```
