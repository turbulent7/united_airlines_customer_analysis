```python
# Convert columns to datetime format
calls_df['call_start_datetime'] = pd.to_datetime(calls_df['call_start_datetime'])
calls_df['agent_assigned_datetime'] = pd.to_datetime(calls_df['agent_assigned_datetime'])
calls_df['call_end_datetime'] = pd.to_datetime(calls_df['call_end_datetime'])

# Calculate HT (Handle Time) and ST (Service Time)
calls_df['HT'] = calls_df['call_end_datetime'] - calls_df['agent_assigned_datetime']
calls_df['ST'] = calls_df['agent_assigned_datetime'] - calls_df['call_start_datetime']

# Display the DataFrame with the new columns
print(calls_df)
```

```python
# Format HT and ST to show only hours, minutes, and seconds (hh:mm:ss)
calls_df['HT'] = calls_df['HT'].apply(lambda x: str(x).split(' ')[-1])
calls_df['ST'] = calls_df['ST'].apply(lambda x: str(x).split(' ')[-1])

# Display the DataFrame with the formatted HT and ST columns
calls_df
```

```python

# Calculate HT (Handle Time) and ST (Service Time) as timedelta
calls_df['HT'] = calls_df['call_end_datetime'] - calls_df['agent_assigned_datetime']
calls_df['ST'] = calls_df['agent_assigned_datetime'] - calls_df['call_start_datetime']

# Convert HT and ST into minutes
calls_df['HT'] = calls_df['HT'].dt.total_seconds() / 60  # This should work since HT is timedelta
calls_df['ST'] = calls_df['ST'].dt.total_seconds() / 60  # This should work since ST is timedelta

# Display the DataFrame with HT and ST in minutes
print(calls_df[['call_start_datetime', 'agent_assigned_datetime', 'call_end_datetime', 'HT', 'ST']])
```
